```python
import requests 

response = requests.get('https://dummyjson.com/users')

print(response.json()['users'][0]['firstName'])
```

```python
tamanho = len(response.json()['users'])
```

```python
usuarios_lista = response.json()['users']
for i in range(tamanho):
    print(f'{usuarios_lista[i]["firstName"]} {usuarios_lista[i]["lastName"]} - idade: {usuarios_lista[i]["age"]}')
    if usuarios_lista[i]["age"] >= 30:
        print('Essa pessoa tem mais de 30!')
    else:
        print('Essa pessoa tem menos de 30!')
```
